```python
import numpy as np
```

```python
import pandas as pd
```

```python
from sklearn.datasets import make_blobs
```

```python
import matplotlib.pyplot as plt
```

```python
# Number of data points
n_samples = 100
```

```python
# Number of features (2D data for easy visualization)
n_features = 2 
```

```python
# Number of blobs or clusters
n_clusters = 3
```

```python
# Create blobs
X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters,cluster_std =5,shuffle =True, random_state=2,)
```

```python
data_points = X
```

```python
cluster_labels =y
```

```python
print(y)
```

```python
# Visualize the blobs
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.title('Generated Blobs')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

Cluster the synthetic dataset using K-means 

```python
from sklearn.cluster import KMeans
```

```python
# Choose the number of clusters(K)
K = 3
```

```python
# Create a k-means model
kmeans = KMeans(n_clusters=K)
```

```python
# Fit the model to my data. The K-means data partitions the data into K clusters based on similarity 
kmeans.fit(X)
```

```python
# Retrieve the cluster labels for each data point
labels = kmeans.labels_
```

```python
print(labels)
```

```python
# Get cluster centre for each cluster. (central point on graph for each cluster)
centroids = kmeans.cluster_centers_
```

Visualize the K-means Clusters 

```python
# Plot the data points with color-coded clusters
plt.scatter(data_points[:, 0], data_points[:, 1], c=labels, cmap='viridis', s=100)

# Plot cluster centers
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200, label='Centroids')

plt.title('K-Means Clustering')
plt.legend()
plt.show()
```
